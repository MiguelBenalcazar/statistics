Lower bound on negative logarithm. The function 1 − y is always less than the function −log[y]. 

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
# Set the number of samples to generate
num_samples = 1000

# Generate a range of x values for plotting
y_values = np.linspace(0,4, 1000)

# Plot the PDF and CDF of the standard normal distribution
plt.figure(figsize=(12, 10))
plt.plot(y_values, -np.log(y_values), label='- log[y]')
plt.plot(y_values, 1 - y_values, label='1 - y')
plt.title(' Lower bound on negative logarithm.')
plt.xlabel('Input,y')
plt.ylabel('Output')
plt.legend()
```
