```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the probability density function (PDF) for the standard normal distribution
def normal_pdf(x):
    return norm.pdf(x)

# Define the cumulative distribution function (CDF) for the standard normal distribution
def normal_cdf(x):
    return norm.cdf(x)

# Define the inverse CDF (Quantile function) for the standard normal distribution
def normal_inv_cdf(u):
    return norm.ppf(u)

# Generate a range of x values for plotting
x_values = np.linspace(-3, 3, 1000)

# Plot the PDF and CDF of the standard normal distribution
plt.figure(figsize=(12, 5))



plt.subplot(1, 2, 1)
plt.plot(x_values, normal_pdf(x_values), label='PDF')
plt.title('Standard Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('Pr(x)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_values, normal_cdf(x_values), label='CDF')
plt.title('Standard Normal Distribution CDF')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()

plt.tight_layout()
plt.show()

# Generate a random sample z* from a uniform distribution over [0, 1]
z_star = np.random.rand()

# Use the inverse CDF to obtain the corresponding x* from the standard normal distribution
x_star = normal_inv_cdf(z_star)

# Display the generated sample and its transformed value
print("Random sample z*:", z_star)
print("Corresponding x*:", x_star)

# Plot the transformation for visualization
plt.figure(figsize=(8, 5))
plt.plot(x_values, normal_cdf(x_values), label='CDF')
plt.scatter([x_star], [z_star], color='red', label='Transformed Sample (x*, z*)')
plt.title('Inverse Transform Sampling: Generating x* from z*')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()
```
