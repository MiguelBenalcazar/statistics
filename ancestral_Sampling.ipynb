```python
import numpy as np
import matplotlib.pyplot as plt

# Define conditional probability distributions Pr(y|x) and Pr(z|y)
def pr_y_given_x(x):
    # Example: Gaussian distribution with mean and standard deviation depending on x
    mean_y = 2 * x
    std_dev_y = 1
    return np.random.normal(loc=mean_y, scale=std_dev_y)

def pr_z_given_y(y):
    # Example: Exponential distribution with a rate parameter depending on y
    rate_z = np.abs(y)
    return np.random.exponential(scale=1/rate_z)

# Set the number of samples to generate
num_samples = 1000

# Step 1: Sample x* from Pr(x)
x_star = np.random.normal(loc=0, scale=1, size=num_samples)

# Step 2: Sample y* from Pr(y|x*)
y_star = np.array([pr_y_given_x(x) for x in x_star])

# Step 3: Sample z* from Pr(z|y*)
z_star = np.array([pr_z_given_y(y) for y in y_star])

# Plot histograms of the generated samples
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(x_star, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Sampled x* from Pr(x)')
plt.xlabel('x')
plt.ylabel('Density')

plt.subplot(1, 3, 2)
plt.hist(y_star, bins=30, density=True, alpha=0.7, color='green')
plt.title('Sampled y* from Pr(y|x*)')
plt.xlabel('y')
plt.ylabel('Density')

plt.subplot(1, 3, 3)
plt.hist(z_star, bins=30, density=True, alpha=0.7, color='orange')
plt.title('Sampled z* from Pr(z|y*)')
plt.xlabel('z')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
